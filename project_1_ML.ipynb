```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv',usecols=['GrLivArea','FullBath','BedroomAbvGr','SalePrice'])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
x=df.iloc[:,0:3]
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
error=mean_squared_error(y_test,y_pred)
```

```python
error
```

```python
score=r2_score(y_test,y_pred)
```

```python
score
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Original Values')
plt.ylabel('Predicted Values')
plt.title('Original Vs Predicted')
plt.show()
```

```python

features = ['GrLivArea', 'FullBath', 'BedroomAbvGr']
new_df = pd.DataFrame({'FullBath':[1000],'GrLivArea':[1],'BedroomAbvGr':[2]})
new_df = new_df[features]
prediction_lr = lr.predict(new_df)
print(f'prediction_lr: {prediction_lr[0]}')
```
